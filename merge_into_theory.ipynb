```python
# What is merge into 
##MERGE INTO is a SQL statement that upserts data — it updates existing rows, inserts new ones, and optionally deletes stale ones, all in a single atomic transaction against a target table.
# The word upsert = update + insert — that's the core idea.


#* Merge into is a sql statement 
#* it is used to perform insert update and delte operations in a single satatement
#* it compares source and target table based on a join key 
#* it applies actions based on match is found or not 
```

```python
# Where will we use merge into 
# Merge into is used whenever we need to apply incremental changes to the target table without overwriting it completely 

# Example:
##    CDC piplelines ############

##     SCD Type 2 ##################

##    Incemental loads ######################

#     Structured streaming with for each batch 
#     def upsert(batch_df, batch_id):
#     DeltaTable.forName(spark, "target") \
#         .merge(batch_df, "target.id = source.id") \
#         .whenMatchedUpdateAll() \
#         .whenNotMatchedInsertAll() \
#         .execute()
# stream.writeStream.foreachBatch(upsert).start()
#  Each micro-batch is merged into the target → real-time upserts

##     Deduplication ##################################

# Raw table has duplicate order_ids from Kafka retries
# MERGE INTO with WHEN NOT MATCHED only → inserts a row only if it doesn't exist
# Guarantees no duplicates in your silver/gold table

```

```python
#When to not use merge into 
# Situation                               Use instead
# Full daily reload of small table          overwrite
# Append-only event logs                    append
# First-time table creation                 CREATE TABLE AS SELECT
# Simple dedup on read                      DISTINCT in SELECT
```

```python
# Disadvance of Merge into
# 1. File Rewrite Cost — The Biggest One
# 2. Slower Than Simple Writes
# 
# (MERGE has to:
# Scan source
# Scan target
# Join both
# Classify every row
# Rewrite affected files)

# 3. Duplicate Key Problem in Source
# 4. View Re-evaluation Problem
# 5. Cannot Insert and Update the Same Key in One Pass
# 6. Performance Degrades Over Time
```

```python
#Merge into strategies 

#1. Upsert (Update + Insert)
#2. Insert Only (Deduplication Guard)
#3. Full CDC (Insert + Update + Delete)
#4. Full Sync (Insert + Update + Delete Stale)
#5. SCD Type 1 (Overwrite History)
#6. SCD Type 2 (Preserve Full History)
# 7. Conditional / Hash-Based Merge (Skip No-Op Updates)
```

```python
#Qus: What is the difference between merge into in Python and SQL?
#ANS: Merge INTO in SQL and doing  a merge/upsert using Python often achieve same business goal.
# --Insert new rows
# -- Update existing rows
# -- Sometimes delete rows

# But the execution model, flexibility and integration interaction are different.
```

```python
#SQL Merge into:
# Declarative, SQL-engine driven, Compact, Easier for DB devlopers

# Python/PySpark Merge
# Python controls the pipeline, Spark builds execution plan, JVM executes job
```

```python
#  Fundamental Difference
```

| SQL MERGE                           | Python/PySpark MERGE                |
| ----------------------------------- | ----------------------------------- |
| Declarative SQL                     | Programmatic API                    |
| Static query                        | Dynamic logic possible              |
| Easier for analysts/DBA             | Easier for data engineers           |
| Best for DB operations              | Best for ETL pipelines              |
| Limited programming constructs      | Full Python control                 |
| Harder to build dynamic merges      | Easy metadata-driven merges         |
| Usually runs inside database engine | Runs through Spark execution engine |
